```python
#TRATAMENTO DE ATRIBUTOS CATEGORICOS

#LabelEncoder  === #tranf dados categoricos strings em numericos

from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_type = LabelEncoder()
label_encoder_Municipality = LabelEncoder()
label_encoder_DistrictName = LabelEncoder()
label_encoder_NearestStation = LabelEncoder()
label_encoder_FrontageIsGreaterFlag = LabelEncoder()
label_encoder_CityPlanning = LabelEncoder()
label_encoder_Period = LabelEncoder()
```

```python

x[:,0] = label_encoder_type.fit_transform(x[:,0])
x[:,1] = label_encoder_Municipality.fit_transform(x[:,1])
x[:,2] = label_encoder_DistrictName.fit_transform(x[:,2])
x[:,3] = label_encoder_NearestStation.fit_transform(x[:,3])
x[:,10] = label_encoder_CityPlanning.fit_transform(x[:,10])
x[:,12] = label_encoder_Period.fit_transform(x[:,12])
```

```python
#padronalização pra deixar os valores na mesma escala, para que o algoritimo nao pense que algum atributo é mais importante do que outro

from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()


for i in range(x.shape[1]):
  # mixed types
  col = x[:, i]
  if any(isinstance(val, str) for val in col) and any(isinstance(val, (int, float)) for val in col):

    x[:, i] = col.astype(str)
  if isinstance(x[0, i], str):
    from sklearn.preprocessing import LabelEncoder
    label_encoder = LabelEncoder()
    x[:, i] = label_encoder.fit_transform(x[:, i])

x = scaler_x.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_treino.shape, x_teste.shape
```

```python
y_treino.shape, y_teste.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_squared_log_error, mean_absolute_percentage_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
```

```python
rf = RandomForestRegressor(n_estimators=200, random_state=42)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

x_treino_imputed = imputer.fit_transform(x_treino)


model = rf.fit(x_treino_imputed, y_treino)
```

```python
imputer = SimpleImputer(strategy='mean')

x_teste_imputed = imputer.fit_transform(x_teste)


previsao = model.predict(x_teste_imputed)
```

```python
r2_score(y_teste, previsao)
```
